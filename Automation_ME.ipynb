```python
df = pd.read_excel(r"C:\Users\nsyrymbet\Desktop\SEB доли.xlsx", header = None)
df
```

```python
df = df.drop(df.columns[[1,2,7]], axis = 1)
df

# df = df.drop(df.columns[[1,2]], axis = 1)
# df
```

```python
df = df.T
df
```

```python
df = df.rename(columns=df.iloc[0])

df.rename(columns={df.columns[0]:'Sales'}, inplace=True)
df.rename(columns={df.columns[1]:'Shops'}, inplace=True)

df = df.drop(df.index[0])

df['Sales'] = df['Sales'].fillna(method='ffill')
df = df.reset_index(drop = True)


df
```

```python
unpivoted_df = pd.melt(df, id_vars=['Sales', 'Shops'], var_name='Category', value_name='Value')
unpivoted_df = unpivoted_df.dropna()
unpivoted_df = unpivoted_df.drop_duplicates()
unpivoted_df = unpivoted_df[unpivoted_df['Value'] != 0]
unpivoted_df = unpivoted_df[unpivoted_df['Value'] != '#DIV/0']
unpivoted_df
```

```python
pivot = pd.pivot_table(unpivoted_df, values='Value', 
                                index=['Sales', 'Category'], 
                                columns='Shops', 
                                aggfunc=np.mean)
pivot
```

```python
pivot = pivot.reset_index()
pivot['Sales'] = pivot['Sales'].fillna(method='ffill')
pivot = pivot.reset_index()
pivot
```

```python
pivot_without_sums = pivot[(pivot['Category'] != 'Итого') & (pivot['Sales'] != 'Итого')]
pivot_without_sums = pivot_without_sums[['Sales', 'Category', 'Общий процент количество', 'Общий процент сумма']]
pivot_without_sums.rename(columns={'Общий процент количество': 'шт', 'Общий процент сумма': 'тг'}, inplace=True)
pivot_without_sums
```

```python
pivot_with_sums = pivot[(pivot['Category'] == 'Итого') & (pivot['Sales'] != 'Итого')]
pivot_with_sums = pivot_with_sums.replace('Итого', 'Итог по  магазину')
pivot_with_sums = pivot_with_sums[['Sales', 'Category', 'Общий процент количество', 'Общий процент сумма']]
pivot_with_sums.rename(columns={'Общий процент количество': 'шт', 'Общий процент сумма': 'тг'}, inplace=True)
pivot_with_sums
```

```python
more = pivot[(pivot['Sales'] == 'Итого')]
more = more.replace('Итого', 'Итоги катег-ии')
more.loc[more['Category'] == 'Общий итог', 'Sales'] = 'Общий итог'
more = more[['Sales', 'Category', 'Итоговый процент количество', 'Итоговый процент сумма']]
more.rename(columns={'Итоговый процент количество': 'шт', 'Итоговый процент сумма': 'тг'}, inplace=True)
more
```

```python
final = pivot_without_sums.append([pivot_with_sums, more]).reset_index(drop=True)
final = final.drop_duplicates()
final['шт'] = final['шт'] /100
final['тг'] = final['тг'] /100

final
```
