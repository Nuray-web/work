```python
import openpyxl
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_excel(r"C:\Users\nsyrymbet\Desktop\Доли Себ 49.xlsx")
df
```

```python
unpivoted_df = pd.melt(df, id_vars=['Категория'], var_name='Shops', value_name='Value')
unpivoted_df = unpivoted_df.dropna()
unpivoted_df = unpivoted_df.drop_duplicates()
unpivoted_df = unpivoted_df[unpivoted_df['Value'] != 0]
unpivoted_df = unpivoted_df[unpivoted_df['Value'] != '#DIV/0']
unpivoted_df
```

```python
unpivoted_df['Sales'] = unpivoted_df['Shops'].str[-7:]

replacements = [('49 неделя ', ''), (' Доля по поставщикам ', ''), ('Доля по поставщикам ', 'итог'), ('тенге', ''), ('шт', ''), ('(', ''), (')', '')]

for x, y in replacements:
    x_escaped = re.escape(x)
    y_escaped = re.escape(y)
    unpivoted_df['Shops'] = unpivoted_df['Shops'].str.replace(x_escaped, y_escaped, regex=True)

unpivoted_df
```

```python
pivot = pd.pivot_table(unpivoted_df, values='Value', 
                                index=['Shops', 'Категория'], 
                                columns='Sales', 
                                aggfunc=np.mean)
pivot
```

```python
pivot = pivot.reset_index()
pivot['Shops'] = pivot['Shops'].fillna(method='ffill')
pivot = pivot.reset_index()
pivot
```

```python
pivot = pivot[['Shops', 'Категория', ' (шт)  ', 'ам (шт)', '(тенге)', 'енге)  ']]
pivot
```

```python
pivot_without_sums = pivot[(pivot['Категория'] != 'Общий итог')]
pivot_without_sums = pivot_without_sums[['Shops', 'Категория', 'ам (шт)', '(тенге)']]
pivot_without_sums = pivot_without_sums[~pivot_without_sums['Shops'].str.contains('итог')]
pivot_without_sums.rename(columns={'ам (шт)': 'шт', '(тенге)': 'тг'}, inplace=True)

pivot_without_sums
```

```python
pivot_with_sums = pivot[pivot['Категория'] == 'Общий итог']
pivot_with_sums = pivot_with_sums.replace('Общий итог', 'Итоги по магазинам')
pivot_with_sums = pivot_with_sums[['Shops', 'Категория', 'ам (шт)', '(тенге)']]
pivot_with_sums.rename(columns={'ам (шт)': 'шт', '(тенге)': 'тг'}, inplace=True)
pivot_with_sums = pivot_with_sums[~pivot_with_sums['Shops'].str.contains('итог')]

pivot_with_sums
```

```python
more = pivot[['Shops', 'Категория', ' (шт)  ', 'енге)  ']]
more = more[more['Shops'].str.contains('итог')]
more.rename(columns={' (шт)  ': 'шт', 'енге)  ': 'тг'}, inplace=True)
more['Shops'] = more['Shops'].str.replace(r'итог', 'Итог по  катег-ии', regex=True)

more.loc[more['Категория'] == 'Общий итог', 'Shops'] = 'Общий итог'

more['Shops'] = more['Shops'].str.strip()

more
```

```python
final = pivot_without_sums.append([pivot_with_sums, more]).reset_index(drop=True)
final = final.drop_duplicates()
final
```

```python
final.to_excel(r"C:\Users\nsyrymbet\Desktop\shares ev.xlsx", index=False)
```
