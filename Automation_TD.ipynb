```python
import openpyxl
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter
import pandas as pd
import numpy as np
```

```python
wb = load_workbook(r"C:\Users\nsyrymbet\Desktop\СЕБ-Восток 3.xlsx")
sheet1 = wb['По неделям']
# sheet1 = wb['Месяц']
sheet2 = wb['По Магазинам']

ws = wb.active

# Добавила лишнюю строку
sheet2.insert_rows(3)

# Я пыталась сделать клир
for row in sheet2['C2:AAT3']:
    for cell in row:
        cell.value = None

# Я добавила первую строку с колонками
names = ['$', 'Share%','Продажи в шт', 'Остатки шт']
for row in sheet2['C2:AAT2']:
    n = 1
    for cell in row:
        if n%4 == 1:
            cell.value = names[0]
        elif n%4 == 2:
            cell.value = names[1]
        elif n%4 == 3:
            cell.value = names[2]
        elif n%4 == 0:
            cell.value = names[3]
        n += 1
        
# добавила вторую строку с магазинами
for row in sheet2['C3:AAX3']:
    n = 3
    m = 1
    for cell in row:
        cell.value = sheet2.cell(row=4, column=n).value
        if m%4 == 0:
            n += 4
        m += 1
        
sheet2.delete_rows(4)

min_row = 2
max_row = 62
min_col = 2
max_col = 723

transposed_sheet = wb.create_sheet('Transposed')

for row in range(min_row, max_row+1):
    for col in range(min_col, max_col+1):
        transposed_sheet.cell(row=col,column=row).value = sheet2.cell(row=row,column=col).value

transposed_sheet.delete_rows(1)
transposed_sheet.delete_cols(1)

week_sheet = wb.create_sheet('Week')

cols = ['Shops','Category','$', 'Share%','Продажи в шт', 'Остатки шт', 'online']
for col in range(1,8):
    week_sheet.cell(row=1, column=col).value = cols[col-1]

for row in range(4, 66):
    for col in range(2, 5):
        week_sheet.cell(row=row-2,column=col).value = sheet1.cell(row=row, column=col).value
        week_sheet.cell(row=row-2,column=7).value = 'TOTAL'

for row in range(68, 130):
    for col in range(2, 5):
        week_sheet.cell(row=row-4,column=col).value = sheet1.cell(row=row, column=col).value
        week_sheet.cell(row=row-4,column=7).value = 'Online'

for row in range(132, 194):
    for col in range(2, 5):
        week_sheet.cell(row=row-6,column=col).value = sheet1.cell(row=row, column=col).value
        week_sheet.cell(row=row-6,column=7).value = 'offline'
        
gen = ['Климат (КБТ)', 'Общий итог МБТ + Климат (КБТ)', 'Пылесосы', 'Общий итог МБТ', 'Посуда', 'Общий итог без посуды и пылесосов']
for row in range(2, 188):
    if week_sheet.cell(row=row, column=2).value in gen:
        week_sheet.cell(row=row, column=1).value = 'Общие итоги'
    else:
        week_sheet.cell(row=row, column=1).value = 'Итоги  кат-ии'
        
wb.save(r"C:\Users\nsyrymbet\Desktop\СЕБ-Восток 3 (wo B2B).xlsx")
```

```python
df = pd.read_excel(r"C:\Users\nsyrymbet\Desktop\СЕБ-Восток 3 (wo B2B).xlsx", sheet_name = 'Transposed')
df.rename(columns={'Unnamed: 0':'Sales'}, inplace=True)
df.rename(columns={'Unnamed: 1':'Shops'}, inplace=True)
df
```

```python
unpivoted_df = pd.melt(df, id_vars=['Sales', 'Shops'], var_name='Category', value_name='Value')
unpivoted_df = unpivoted_df.dropna()
unpivoted_df = unpivoted_df.drop_duplicates()
unpivoted_df = unpivoted_df[unpivoted_df['Value'] != 0]
unpivoted_df = unpivoted_df[unpivoted_df['Value'] != '#DIV/0']
unpivoted_df
```

```python
pivot = pd.pivot_table(unpivoted_df, values='Value', 
                                index=['Shops', 'Category'], 
                                columns='Sales', 
                                aggfunc=np.mean)
pivot
```

```python
pivot = pivot.reset_index()
pivot['Shops'] = pivot['Shops'].fillna(method='ffill')
pivot
```

```python
pivot = pivot[['Shops','Category','$', 'Share%','Продажи в шт', 'Остатки шт']]
pivot
```

```python
pivot_without_sums = pivot[pivot['Category'] != 'Сумма:']
pivot_without_sums
```

```python
pivot_with_sums = pivot[pivot['Category'] == 'Сумма:']
pivot_with_sums = pivot_with_sums.replace('Сумма:', 'Итоги по магазину')
pivot_with_sums
```

```python
more = pd.read_excel(r"C:\Users\nsyrymbet\Desktop\СЕБ-Восток 3 (wo B2B).xlsx", sheet_name = 'Week')
more
```

```python
final = pivot_without_sums.append([pivot_with_sums, more]).reset_index(drop=True)
final = final.drop_duplicates()
final
```

```python
final.to_excel(r"C:\Users\nsyrymbet\Desktop\shares td.xlsx", index=False)
```
