```python
import openpyxl
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter
import pandas as pd
import numpy as np
import re
```

```python
link = r"C:\Users\nsyrymbet\Desktop\Поставщикам (with Shares).xlsx"

wb = load_workbook(link)
total = wb['TOTAL']

total.delete_rows(1)

wb.save(link)
```

```python
df = pd.read_excel(link, sheet_name = 'TOTAL')
df
```

```python
df = df.fillna(method="ffill", axis=1)
df
```

```python
df = df.drop(columns = ['Business Team\n', 'Level 2', 'Level 3'])
df
```

```python
df['Shops'] = pd.Series(["Итог по категориям" for x in range(len(df.index))])
df
```

```python
df = df[['Shops', 'Level 5', 'Amount Share', 'Qty Share']]
df = df.dropna()
df = df.drop_duplicates()
df
```

```python
df = df.dropna()
df = df.drop_duplicates()
df = df[(df['Amount Share'] != 0) & (df['Qty Share'] != 0)]
df = df[(df['Amount Share'] != '#DIV/0') & (df['Qty Share'] != '#DIV/0')]
df = df.rename(columns = {'Level 5': 'Category', 'Amount Share':'Value'})
df
```

```python
link2 = r"C:\Users\nsyrymbet\Desktop\Отчет по долям (магазины группы).xlsx"

wb = load_workbook(link2)
sheet = wb['Отчет по долям (магазины группы']

sheet.delete_rows(1)
sheet.delete_rows(1)
sheet.delete_rows(2)

wb.save(link2)

df2 = pd.read_excel(link2, sheet_name = 'Отчет по долям (магазины группы')
df2
```

```python
df3 = df2[['Unnamed: 0', 'Unnamed: 1']].fillna(method="ffill", axis=1)
df3
```

```python
df2['Unnamed: 1'] = df3['Unnamed: 1']
df2 = df2.drop(columns =['Unnamed: 0'])
df2
```

```python
unpivoted_df = pd.melt(df2, id_vars=['Unnamed: 1'], var_name='Shops', value_name='Value')
unpivoted_df = unpivoted_df.dropna()
unpivoted_df = unpivoted_df.drop_duplicates()
unpivoted_df = unpivoted_df[unpivoted_df['Value'] != 0]
unpivoted_df = unpivoted_df[unpivoted_df['Value'] != '#DIV/0!']
unpivoted_df = unpivoted_df.rename(columns = {'Unnamed: 1': 'Category'})
unpivoted_df = unpivoted_df[['Shops', 'Category', 'Value']]
unpivoted_df
```

```python
df_without_sums = unpivoted_df[(unpivoted_df['Category'] != 'Итого:')]
df_without_sums
```

```python
df_with_sums = unpivoted_df[(unpivoted_df['Category'] == 'Итого:')]
df_with_sums = df_with_sums.replace('Итого:', 'Итог по магазину')
df_with_sums
```

```python
final = df_without_sums.append([df_with_sums, df]).reset_index(drop=True)
final = final.drop_duplicates()
final
```

```python
final.to_excel(r"C:\Users\nsyrymbet\Desktop\shares su.xlsx", index=False)
```
